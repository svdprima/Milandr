```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm

data = np.loadtxt ("data.csv",  dtype='float', delimiter=';')
s1 = []
s2 = []
z = []
for i in data:
    if i[0] not in s1:
        s1.append(i[0])
    elif i[1] not in s2:
        s2.append(i[1])
    z.append(i[2])
Z = np.zeros (shape = (len(s1), len(s2)))
for i in range (0,len(s1)):
    for j in range (0,len(s2)):
        Z[i][j] = z [i * (len (s2) + 1) + j]
print (Z)
S1, S2 = np.meshgrid(s2, s1)
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(S1, S2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('S2')
ax.set_ylabel('S1')
ax.set_zlabel('U_diff')
```
